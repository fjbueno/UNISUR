```python
import plotly.graph_objects as go
import plotly.offline as py
import pandas as pd
from plotly.offline import download_plotlyjs, init_notebook_mode, plot,iplot
import plotly.express as px

df = pd.read_csv('enfermedades2020mes2.csv')
fig = px.scatter(df, x="especialidad", y="edad", color="sexo",
                labels={
                     "especialidad": "Especialidad",
                     "edad": "Edad",
                     "sexo": "Sexo"
                 }, 
                title="Proyeccin de Expedientes por Especialidad, Sexo y Edad")

fig.update_layout(
    font_family="Courier New",
    font_color="blue",
    title_font_family="Times New Roman",
    title_font_color="black",
    legend_title_font_color="green"
)
fig.update_xaxes(title_font_family="Arial")
fig.show()
py.plot(fig, filename='grafico5.html')
```

```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.offline as py
import pandas as pd
from plotly.offline import download_plotlyjs, init_notebook_mode, plot,iplot

df = pd.read_csv('enfermedades2020mes2.csv')
df.set_index('especialidad')

fig = px.box(df, x="especialidad", y="edad" , color="especialidad",
            labels={
                     "especialidad": "Especialidad",
                     "edad": "Edad",
                    }, 
                title="Proyeccin de Atencin:Especialidad por Edad"
            )
fig.update_layout(
    font_family="Courier New",
    font_color="blue",
    title_font_family="Times New Roman",
    title_font_color="black",
    legend_title_font_color="green"
)
fig.update_xaxes(title_font_family="Arial")
fig.update_traces(quartilemethod="exclusive") # or "inclusive", or "linear", or "exclusive"  by default
fig.show()
py.plot(fig, filename='Grafico2.html')
```

```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.offline as py
import pandas as pd
from plotly.offline import download_plotlyjs, init_notebook_mode, plot,iplot

df = pd.read_csv('enfermedades2020mes2.csv')
df.set_index('sexo')
fig = px.box(df, x="mes", y="especialidad", color="sexo",
             notched=True, # used notched shape
             # title="Box plot of total bill",
             hover_data=["sexo"], # add day column to hover data
             labels={
                     "mes": "Atencin a los Pacientes por Mes",
                     "especialidad": "Especialidades Demandadas",
                    }, 
                title="Proyeccin de Atencin: Especialidad por Mes y Sexo"
            )
            
fig.show()
py.plot(fig, filename='jupyter-3box.html')
```
